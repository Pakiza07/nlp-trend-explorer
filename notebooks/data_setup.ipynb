# Dataset Setup for NLP Trend Explorer

**Objective:**  
- Select a dataset (news headlines)  
- Keep only relevant columns (`text`, `date`)  
- Clean basic structure for preprocessing 
- Save a CSV as base dataset for the project  

This notebook sets up the foundation for all subsequent NLP and trend analysis tasks.

## Dataset Description

- **Source:** News headlines dataset   
- **Columns of Interest:** `description` (text), `published_at` (date)  
- **Rows:** 553 (subset for speed)  
- **Goal:** Standardize columns and save a cleaned CSV (`base_data.csv`)

```python
import pandas as pd
```

```python
#Load dataset
df = pd.read_csv("C:\\Users\\ASUS\\nlp-trend-explorer\\data\\raw\\cnbc_news_data.csv") 

df.head()
```

```python
# Keep only text and date
df = df[['description', 'published_at']] 
df.columns = ['text', 'date']

# Drop empty rows
df = df.dropna(subset=['text']) 

df = df.head(1500)

df.head()
```

```python
df.to_csv("../data/raw/base_data.csv", index=False)
```
